```python
# 1. Base Facts (Subject, Relation, Object)
kb_facts = [
    ("Nairobi", "is_in", "Kenya"),
    ("Kenya", "is_in", "Africa"),
    ("Africa", "is_on", "Earth")
]

# 2. Rule: If X is in Y, and Y is in Z, then X is in Z
inferred_facts = []
for f1 in kb_facts:
    for f2 in kb_facts:
        if f1[2] == f2[0]: # If the object of one matches the subject of another
            inferred_facts.append((f1[0], "is_ultimately_in", f2[2]))

# 3. Display Results
print("--- Base Facts ---")
for f in kb_facts: print(f)

print("\n--- Inferred Facts ---")
for f in inferred_facts: print(f)

# 4. Query: Where is Nairobi?
query = [f[2] for f in inferred_facts if f[0] == "Nairobi"]
print("\nQuery Result: Nairobi is ultimately in {}".format(query))
```
