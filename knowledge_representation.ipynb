```python
# Facts represented as (subject, relation, object)

facts = [
    ("John", "is_a", "Student"),
    ("Mary", "is_a", "Student"),
    ("Student", "can", "Study"),
    ("Study", "leads_to", "Knowledge")
]

facts
```

```python
# Rule 1:
# If X is_a Student and Student can Study
# Then X can Study

inferred_facts = []

for fact1 in facts:
    for fact2 in facts:
        if fact1[1] == "is_a" and fact2[0] == fact1[2] and fact2[1] == "can":
            inferred_facts.append((fact1[0], "can", fact2[2]))

inferred_facts
```

```python
# Rule 2:
# If X can Study and Study leads_to Knowledge
# Then X gains Knowledge

for fact in inferred_facts:
    for f in facts:
        if fact[2] == f[0] and f[1] == "leads_to":
            inferred_facts.append((fact[0], "gains", f[2]))

inferred_facts
```

```python
# Query: Who gains Knowledge?

for fact in inferred_facts:
    if fact[1] == "gains" and fact[2] == "Knowledge":
        print(f"{fact[0]} gains Knowledge")
```
